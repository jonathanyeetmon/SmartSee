```python
# Weights will be auto-downloaded
import cv2
from predictor import predictor

image_paths = ['images/Nike_tshirt.jpeg','images/Adidas_tshirt.jpg','images/Hang_Laptop.png','images/Anirudh_nike.JPG',
              'images/Sunny_dell.jpg','images/Adidas_tshirt_2.jpg']

predictions = predictor(image_paths)
```

```python
print(predictions)
```

```python
import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle
import numpy
from PIL import Image
```

```python
for i in range(len(predictions)):
    plt.imshow(numpy.array(Image.open(image_paths[i])))
    plt.gca().add_patch(Rectangle((predictions[i][0]['box'][0],predictions[i][0]['box'][1]),predictions[i][0]['box'][2]-predictions[i][0]['box'][0],predictions[i][0]['box'][3]-predictions[i][0]['box'][1],linewidth=1,edgecolor='g',facecolor='none'))
    plt.show()
    print(predictions[i][0]['label'])
    print(f'Confidence: {predictions[i][0]["score"]*100}')
    print('\n\n\n')
```
